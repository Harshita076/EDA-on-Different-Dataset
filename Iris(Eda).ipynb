```python
#Scatter PLot
sns.scatterplot(x="SepalLengthCm",y="SepalWidthCm",hue="Species", data=df)
plt.show()
```

```python
#Line Chart
sns.lineplot(x="PetalLengthCm", y="PetalWidthCm", data=df)
plt.show()
```

```python
#Relational PLot(relplot)
sns.relplot(x="SepalLengthCm", y="SepalWidthCm",col="Species", data=df)
plt.show()
```

```python
#Histogram
sns.histplot(df["SepalLengthCm"], bins=50, kde=False)
plt.show()
```

```python
#KDE PLot (kernel density estimate)
sns.kdeplot(df["SepalWidthCm"], fill=True)
plt.show()
```

```python
#KDE+ histogram 
sns.histplot(df["SepalWidthCm"],bins=20,kde=True)
plt.show()
```

```python
#Bar Chart
sns.barplot(x="Species", y="PetalLengthCm", data=df)
plt.show()
```

```python
#Boxplot
sns.boxplot(x="Species", y="PetalWidthCm", data=df)
```

```python
#Violin Plot
sns.violinplot(x="Species",y="SepalLengthCm",data=df)
plt.title("Sepal Length Distribution by Species")
plt.show()
```

```python
#Heatmap    #Here Red colour represents- positive corerelation whereas blue colour shows negative corerelation 
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
#Pair plot
sns.pairplot(df, hue="Species")
plt.show()
```

```python
#Countbar 
#Countbar 
sns.countplot(x="Species", data=df)
plt.show()
```
